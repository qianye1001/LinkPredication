```python
import json
from Tools.CreateGraph import CreateGraph
import networkx as nx
from Tools.GetEdges import GetEdges

print("读取数据")
test_edges = GetEdges(r"..\Date\test_date.txt",109580)
non_edges = GetEdges(r"..\Date\non_edges.txt",976560)

# test_edges = GetEdges(r"..\Date\test_date.txt",1090)
# non_edges = GetEdges(r"..\Date\non_edges.txt",9765)

G = CreateGraph("../Date/课程设计数据集.txt").GetNx()
G.remove_edges_from(test_edges)
with open("../Date/NC_Mod.json",'r',encoding='utf8') as f:
    comm = json.load(f)
    print(comm)

def My(node1, node2):
    if comm[str(node1)] != comm[str(node2)]:
        jc = nx.jaccard_coefficient(G, [(node1,node2)])
        return next(jc)[2]
    elif nx.shortest_path_length(G, source=node1, target=node2) > 3:
        return 0
    else:
        score = 0
        for nbr in list(G.adj[node1]):
            if comm[str(node1)] == comm[str(nbr)] and nbr != node2:
                for nbr2 in list(G.adj[nbr]):
                    if nbr2 != node2 and comm[str(node1)] == comm[str(nbr2)] and G.has_edge(nbr2,node2):
                        score = score + 1/G.degree[nbr] + 1/G.degree[nbr2]
        a = len(set(list(G.adj[node1]) +list(G.adj[node2])))
        return (a + 0.1 * score)/(len(list(G.adj[node1])) + len(list(G.adj[node2])) + 0.1 * score)

print("计算Cja开始")
jc_t = [My(x,y) for x,y in test_edges]
jc_n = [My(x,y) for x,y in non_edges]
print("计算Cja结束")

print("计算AUC")
count = 0
score = 0
for n in jc_n:
    for t in jc_t:
        count = count + 1
        if t > n:
            score = score + 1
        elif t == n:
            score = score + 0.5
print(score/count)

```
